```python
import wandb

# Initialize the API
api = wandb.Api()

# Define your project details
entity = "kevinmathew"  # Your wandb username or team
project = "Prostate MRI Classification"  # Your wandb project name

# Fetch all runs in the project
runs = api.runs(f"{entity}/{project}")

for run in runs:
    print(f"Processing Run: {run.id}")

    # List all files in the run
    files = run.files()

    for file in files:
        if "weights/" in file.name or file.name.startswith("weights"):  
            print(f"Deleting: {file.name} from Run {run.id}")
            try:
                file.delete()  # Delete the file remotely
            except Exception as e:
                print(f"Error deleting {file.name}: {e}")

print("All 'weights' folders deleted successfully from WandB.")
```
